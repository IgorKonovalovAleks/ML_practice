```python
a = 0
for col in df.columns[:-1]:
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].quantile(0.75)
    IQR = Q2 - Q1
    l = Q1 - 1.5 * IQR
    u = Q2 + 1.5 * IQR
    print(((df[col] < l) | (df[col] > u)).value_counts())
a
```

В выборке довольно много выбросов, лучше от них избавиться
Более экстремальными значениями обладают места с плохим качеством воздуха, поэтому лучше расширить границы допустимых значений, потому что места с плохим качеством воздуха будут отсеиваться непропорционально часто

```python
a = 0
data = df.copy()
for col in data.columns[:-1]:
    Q1 = data[col].quantile(0.20)
    Q2 = data[col].quantile(0.85)
    IQR = Q2 - Q1
    l = Q1 - 1.5 * IQR
    u = Q2 + 1.5 * IQR
    rows_to_drop = data[(data[col] < l) | (data[col] > u)].index
    data = data.drop(rows_to_drop)
    print(data.shape)
data['Air Quality'].value_counts()
```

теперь разобьем данные на обучающую и тестовую выборки

```python
X = data.drop(['Air Quality'], axis=1)
Y = data['Air Quality']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)

N_train, d = X_train.shape
N_test,  d = X_test.shape

N_train, N_test, d
```

Проведем стандартизацию значений, то есть приведем все значения к нулевому среднему и единичному стандартному отклонению
Это нужно провести перед началом работы алгоритмов машинного обучения

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns, index=X_train.index)

scaler = StandardScaler()
scaler.fit(X_test)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=7, p=1)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
classes = list(pd.Categorical(data['Air Quality']).categories)

cm = confusion_matrix(y_test, y_test_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot()
```

Как видно из матрицы рассогласования, чаще всего модель ошибается, завышая качество на один уровень. Оно неудивительно, так как большинство записей кучкуются по нижней границе измеряемых значений, что можно было наблюдать из диаграмм. так же видно, что модель хуже предсказывает худшее качество воздуха, чем лучшее. Так происходит из-за несбалансированности классов: мест с хорошим качеством воздуха гораздо больше, чем с плохим, к тому же, хороший воздух по составу примерно одинаков, когда как значения для плохого могут быть очень разными.

```python
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3, p=1)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print(classification_report(y_test, y_test_pred))
```
